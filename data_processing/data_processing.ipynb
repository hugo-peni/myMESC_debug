```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from data.txt with skipinitialspace to handle spaces after commas
data = pd.read_csv('data.txt', skipinitialspace=True)

# Create two separate plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Plot 1: Three Phase Currents
ax1.plot(data['Ia'], label='Phase A Current', linewidth=1.5)
ax1.plot(data['Ib'], label='Phase B Current', linewidth=1.5)
ax1.plot(data['Ic'], label='Phase C Current', linewidth=1.5)
ax1.set_xlabel('Sample Number', fontsize=12)
ax1.set_ylabel('Current (A)', fontsize=12)
ax1.set_title('Three Phase Current Values', fontsize=14, fontweight='bold')
ax1.legend(loc='best', fontsize=10)
ax1.grid(True, alpha=0.3)

# Plot 2: Three Phase Voltages
ax2.plot(data['Va'], label='Phase A Voltage', linewidth=1.5)
ax2.plot(data['Vb'], label='Phase B Voltage', linewidth=1.5)
ax2.plot(data['Vc'], label='Phase C Voltage', linewidth=1.5)
ax2.set_xlabel('Sample Number', fontsize=12)
ax2.set_ylabel('Voltage (V)', fontsize=12)
ax2.set_title('Three Phase Voltage Values', fontsize=14, fontweight='bold')
ax2.legend(loc='best', fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
